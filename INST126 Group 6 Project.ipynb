```python
import pandas as pd
import matplotlib as plt
%pylab inline

covid1 = pd.read_csv('04-12-2020.csv')
covid2 = pd.read_csv('04-23-2020.csv')

usa_whole_confirmed_12 = 0
usa_whole_deaths_12 = 0
usa_mort_rate_12 = 0
usa_case_rate_12 = 0
state_count_12 = 0
state_counts_12 = 0

usa_whole_confirmed_23 = 0
usa_whole_deaths_23 = 0
usa_mort_rate_23 = 0
usa_case_rate_23 = 0
state_count_23 = 0
state_counts_23 = 0

for x in covid1['Confirmed']:
    usa_whole_confirmed_12 += x

for x in covid1['Deaths']:
    usa_whole_deaths_12 += x

for x in covid1['Mortality_Rate']:
    if x > 0:
        state_count_12 += 1
        usa_mort_rate_12 += x

usa_mort_rate_12 /= state_count_12

for x in covid1['Incident_Rate']:
    if x > 0:
        state_counts_12 += 1
        usa_case_rate_12 += x

usa_case_rate_12 /= state_counts_12

# Below: 4-23 data

for x in covid2['Confirmed']:
    usa_whole_confirmed_23 += x

for x in covid2['Deaths']:
    usa_whole_deaths_23 += x

for x in covid2['Mortality_Rate']:
    if x > 0:
        state_count_23 += 1
        usa_mort_rate_23 += x

usa_mort_rate_23 /= state_count_23

for x in covid2['Incident_Rate']:
    if x > 0:
        state_counts_23 += 1
        usa_case_rate_23 += x

usa_case_rate_23 /= state_counts_23

april_12_23 = {'Dates': ['04/12', '04/23'], 'Confirmed Cases': [usa_whole_confirmed_12, usa_whole_confirmed_23], 'Death Toll': [usa_whole_deaths_12, usa_whole_deaths_23], 'Incident Rate': [usa_case_rate_12, usa_case_rate_23], 'Death Rate': [usa_mort_rate_12, usa_mort_rate_23]}

cvd19 = pd.DataFrame(april_12_23, columns= ['Dates', 'Confirmed Cases', 'Death Toll', 'Incident Rate', 'Death Rate'], index=[0, 1])

plt.bar(cvd19['Dates'], cvd19['Confirmed Cases'], color='darkgreen')
plt.title('Confirmed Cases in the US on 4/12 vs on 4/23')
plt.xlabel('Date')
plt.ylabel('Number of Confirmed Cases')
```

```python
plt.bar(cvd19['Dates'], cvd19['Death Toll'], color='black')
plt.title('Death Toll in the US on 4/12 vs on 4/23')
plt.xlabel('Date')
plt.ylabel('Number of Deaths')
```

```python
plt.bar(cvd19['Dates'], cvd19['Incident Rate'], color='firebrick')
plt.title('Incident Rate in the US on 4/12 vs on 4/23')
plt.xlabel('Date')
plt.ylabel('Incident Rate')
```

```python
plt.bar(cvd19['Dates'], cvd19['Death Rate'], color='black')
plt.title('Death Rate in the US on 4/12 vs on 4/23')
plt.xlabel('Date')
plt.ylabel('Death Rate')
```
